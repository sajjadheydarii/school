# **ŸêData Visualisation**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Input
years = [2015,2016,2017,2018,2019,2020,2021,2022]
Iran_dollar = [4,3.5,3,6,9,20,25,27]
KSA_dollar = [2,2.5,3,3.2,4,4,5,6]

#Making Plots
fig = plt.figure(figsize = (10,10))
plt.plot(years, Iran_dollar, color = "red")
plt.plot(years, KSA_dollar, color = "darkgreen", linestyle = "--")

#Some methods for better visualising
plt.xlabel("year")
plt.ylabel("dollar amount (calibrated in thousand tomans)")
plt.title("Iran vs KSA dollar difference in various years")
plt.legend(["Iran", "KSA"])
plt.yticks([j for j in range(1,32,3)])

#Annotate chart
plt.annotate("Iran and 5 + 1 negotiation",
             xytext = (2016,10),
             xy = (2017,3),
             fontsize = 13,
             arrowprops = dict(facecolor = "black", width = 2))

#Additional Codes
plt.grid()
plt.show()
```

```python
days = [j for j in range(1,11)]
rain_values = [5,266,208,126,115,81,79,144,67,399]

fig = plt.figure(figsize = (8,8))
plt.bar(days, rain_values, color = "lightcoral")

plt.title("Rain quantity in 10 past days of a special city")
plt.xlabel("day")
plt.ylabel("rain value")
plt.xticks([k for k in range(1,11)])

plt.show()
```

```python
days = [j for j in range(1,11)]
rain_values = [5,266,208,126,115,81,79,144,67,399]

fig = plt.figure(figsize = (8,8))
plt.barh(days, rain_values, color = "crimson")

plt.title("Rain quantity in 10 past days of a special city")
plt.ylabel("day")
plt.xlabel("rain value")
plt.yticks([k for k in range(1,11)])

plt.show()
```

```python
x = np.random.normal(0,1,1024)
y = np.random.normal(0,1,1024)

fig = plt.figure(figsize = (8,8))
plt.scatter(x,y, color = "teal", s = 10)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Normal Distribution Scatter Plot")

plt.show()
```
